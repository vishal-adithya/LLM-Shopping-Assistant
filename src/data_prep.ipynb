# Data Preprocessing 

```python
import os
import pandas as pd
```

```python
CATEGORIES_DATASET = os.path.join('../Data', 'amazon_categories.csv')
PRODUCTS_DATASET = os.path.join('../Data', 'amazon_products.csv')

categories_df = pd.read_csv(CATEGORIES_DATASET)
products_df = pd.read_csv(PRODUCTS_DATASET)

products_df.tail()
```

```python
categories_df.head()
categories_df["category_id"] = categories_df["id"]
categories_df.drop(columns=["id"], inplace=True)
categories_df.head()
```

```python
merged_df = products_df.merge(categories_df, on="category_id", how="left")
merged_df.head()
```

```python
merged_df["product_id"] = merged_df["asin"].astype(str)
merged_df.drop(columns=["asin","imgUrl","productURL"], inplace=True)
```

```python
PREPROCESSED_DATASET = os.path.join('../Data', 'preprocessed_data.csv')

merged_df.to_csv(PREPROCESSED_DATASET, index=False)
merged_df.head()
```
